### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1. Object-Oriented (OO) Interface (Explicit Interface)
In the object-oriented approach, you explicitly create Figure and Axes objects and call methods on them to create the plots. This approach gives more control over the plot and is generally preferred when you want to create more complex plots or when you need to manage multiple subplots.
2. Pyplot Interface (Implicit Interface)
In the pyplot interface, you use functions from matplotlib.pyplot to create and manipulate plots. pyplot automatically manages the creation of Figure and Axes objects, making it simpler for basic plots. This is useful for quickly creating simple plots.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: The figure is the whole figure, the overall container for everything you see in a plot, the canvas on which all the plotting elements are drawn. It can contain one or more subplots (axes), titles, legends, and other elements.

Axes: Axes are the individual plots inside the figure, a region for plotting data. Each Axes object contains the data, x- and y-axis, labels, title, etc.

Axis: The axis refers to the x-axis or y-axis of an Axes object. Each Axes has two axes (horizontal and vertical) that control the ticks, limits, and labels.

Artist: An Artist is everything you see on the plot: lines, text, shapes, or any visual element. Both Figure and Axes are containers for artists, and every object drawn on the figure (e.g., plots, labels) is an artist.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 9)
y = x**2

# OO-approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Created with the object-oriented approach")

# pyplot-approach
plt.figure()
plt.plot(x, y)
plt.title("Created with the pyplot approach")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(2, 1, layout="constrained") 

# Scatter plot
axs[0].scatter(random_data_x, random_data_y, c = "blue", alpha = 0.5)
axs[0].scatter(x, y, color = "red", linewidth=2)
axs[0].set_title("Scatter plot")

# Bar plot
axs[1].bar(names, values, color = "blue")
axs[1].set_title("Bar plot")
```
